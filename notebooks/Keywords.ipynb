```python
import pandas as pd

df = pd.read_csv("../links.csv")
df.dropna(subset=["title", "content"], inplace=True, how="all")
df["title_content"] = df["title"] + df["content"]
df.dropna(subset=["title_content"], inplace=True)

texts = df["title_content"].values.tolist()

texts[0:10]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

tfidf = TfidfVectorizer(strip_accents='ascii', lowercase=True, ngram_range=(1, 3), max_df=0.1,
                min_df=5,
                use_idf=True,
                stop_words=stopwords)
tfidf.fit_transform(texts)

def tfidf_keywords(text):
    response = tfidf.transform([text])
    feature_array = np.array(tfidf.get_feature_names())

    tfidf_sorting = np.argsort(response.toarray()).flatten()[::-1]

    if len(text.split(" ")) < 10:
        return ""
    
    n = 10
    top_n = feature_array[tfidf_sorting][:n]
    return " ".join(top_n)

tfidf_keywords(text)
```
